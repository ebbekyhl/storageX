```python
# Import packages
import pandas as pd
import numpy as np
import pypsa
# from six import iteritems
import os
import warnings
warnings.filterwarnings('ignore')

override_component_attrs = pypsa.descriptors.Dict({k : v.copy() for k,v in pypsa.components.component_attrs.items()})
override_component_attrs["Link"].loc["bus2"] = ["string",np.nan,np.nan,"2nd bus","Input (optional)"]
override_component_attrs["Link"].loc["bus3"] = ["string",np.nan,np.nan,"3rd bus","Input (optional)"]
override_component_attrs["Link"].loc["bus4"] = ["string",np.nan,np.nan,"4th bus","Input (optional)"]
override_component_attrs["Link"].loc["efficiency2"] = ["static or series","per unit",1.,"2nd bus efficiency","Input (optional)"]
override_component_attrs["Link"].loc["efficiency3"] = ["static or series","per unit",1.,"3rd bus efficiency","Input (optional)"]
override_component_attrs["Link"].loc["efficiency4"] = ["static or series","per unit",1.,"4th bus efficiency","Input (optional)"]
override_component_attrs["Link"].loc["p2"] = ["series","MW",0.,"2nd bus output","Output"]
override_component_attrs["Link"].loc["p3"] = ["series","MW",0.,"3rd bus output","Output"]
override_component_attrs["Link"].loc["p4"] = ["series","MW",0.,"4th bus output","Output"]
override_component_attrs["StorageUnit"].loc["p_dispatch"] = ["series","MW",0.,"Storage discharging.","Output"]
override_component_attrs["StorageUnit"].loc["p_store"] = ["series","MW",0.,"Storage charging.","Output"]

c_hat = 9.6
c1 = 350
c2 = 339

path = '/home/au485969/project/pypsa-eur-0.3.0/pypsa-eur-sec/results/3H_run_3888/postnetworks'
networks_list = os.listdir(path)
```

```python
df_store = pd.DataFrame()
df_link = pd.DataFrame()
df_onwind = pd.DataFrame()
df_offwind = pd.DataFrame()
df_solar = pd.DataFrame()
df_ror = pd.DataFrame()

df_onwind_cap = pd.DataFrame()
df_offwind_cap = pd.DataFrame()
df_solar_cap = pd.DataFrame()
df_ror_cap = pd.DataFrame()
df_gas_cap = pd.DataFrame()
df_hydro_cap = pd.DataFrame()

df_hydro = pd.DataFrame()
df_gas = pd.DataFrame()
df_battery_loadshift_g = pd.DataFrame()
df_battery_loadshift_l = pd.DataFrame()
df_storage_loadshift_g = pd.DataFrame()
df_storage_loadshift_l = pd.DataFrame()
df_AClines_ur_mean = pd.DataFrame()
df_DClinks_ur_mean = pd.DataFrame()
df_AClines_ur_var = pd.DataFrame()
df_DClinks_ur_var = pd.DataFrame()
j = 0
i = 2
for filename in networks_list[0:2]:
    n = pypsa.Network(path + '/' + filename,
                        override_component_attrs=override_component_attrs)
    sector_i = []
    for i in filename.split('/')[-1].split('-'):
        print(i)
        if 'H2 Store' in i:
            print('cost factor:')
            print(i.split('+')[1][1:])
            cost_factor = float(i.split('+')[1][1:])
        if 'H2 Electrolysis+c' in i:
            print('charge cost factor:')
            print(i.split('+')[1][1:])
            charge_cost_factor = float(i.split('+')[1][1:])
        if 'H2 Fuel Cell+c' in i:
            discharge_cost_factor = float(i.split('+')[1][1:])
        if 'elec' in i:
            nnodes = int(i.split('_')[2][1:])
            trans = float(i.split('_')[3][2:])
            co2_cap = float(i.split('_')[5][4:])

        if i == 'H' or i == 'T' or i == 'B' or i == 'I':
            sector_i.append(i)

    tres = int(filename.split('/')[-1].split('-')[1][0:-1])

    energy_storage_cost = c_hat*cost_factor
    energy_capacity = n.stores[n.stores.carrier == 'H2'].e_nom_opt.sum()
    co2_price = n.global_constraints.at["CO2Limit","mu"]
    discharge_efficiency =  n.links[n.links.carrier == 'H2 Fuel Cell'].efficiency[0]
    charge_efficiency =  n.links[n.links.carrier == 'H2 Electrolysis'].efficiency[0]
    discharge_capacity =  n.links[n.links.carrier == 'H2 Fuel Cell'].p_nom_opt.sum() # discharging
    charge_capacity = n.links[n.links.carrier == 'H2 Electrolysis'].p_nom_opt.sum()
    storage_discharge_t = -n.links_t.p1[n.links[n.links.carrier == 'H2 Fuel Cell'].index].sum().sum() # MWe (after conversion to electricity)
    load_shifting =  storage_discharge_t.sum()/n.loads_t.p.sum().sum()
    standing_loss = n.stores[n.stores.carrier == 'H2'].standing_loss[0]
    tau = (-(np.log(1-standing_loss)**-1)/24).round(1)
    lh_avg = energy_capacity/(n.loads_t.p.sum(axis=1).mean()) # Average load hours
    P = n.links[n.links.carrier == 'H2 Fuel Cell'].p_nom_opt.sort_values()
    E = n.stores[n.stores.carrier == 'H2'].e_nom_opt.sort_values()
    P_mean = P.mean()
    sigma_P_rel = np.std(P)/P_mean*100
    E_mean = E.mean()
    sigma_E_rel = np.std(E)/E_mean*100
    P_max = P_mean + P_mean # Statistical max
    E_max = E_mean + E_mean # Statistical max
    n_P_out = len(P[P > P_max]) # number of outliers (Outlier indicate hydrogen is con)
    n_E_out = len(E[E > E_max]) # number of outliers
    system_cost = n.objective/1e9
    avail = n.generators_t.p_max_pu.multiply(n.generators.p_nom_opt).sum().groupby(n.generators.carrier).sum()
    used = n.generators_t.p.sum().groupby(n.generators.carrier).sum()
    avail_offwind = avail.loc['offwind-ac'] + avail.loc['offwind-dc']
    avail_onwind = avail.loc['onwind']
    avail_solar = avail.loc['solar'] + avail.loc['solar rooftop'] 
    used_offwind = used.loc['offwind-ac'] + used.loc['offwind-dc'] 
    used_onwind = used.loc['onwind']
    used_solar = used.loc['solar'] + used.loc['solar rooftop']
    battery_storage_discharge_t = -n.links_t.p1[n.links[n.links.carrier == 'battery discharger'].index].sum().sum() 
    battery_load_shifting =  battery_storage_discharge_t.sum()/n.loads_t.p.sum().sum()  
    charge_cost = c1*charge_cost_factor
    discharge_cost = c2*discharge_cost_factor
    gen = {}
    test = pd.DataFrame()
    test[0] = n.generators.carrier
    techs = ['onwind','offwind','solar','ror']
    for tech in techs:
        test[1] = [tech in n.generators.carrier[s] for s in range(len(n.generators))]
        gen[tech] = np.round(n.generators_t.p[n.generators[test[1]].index].sum().sum(),1)
    gen_gas = np.round(-n.links_t.p1[n.links[n.links.carrier == 'OCGT'].index].sum().sum(),1)
    gen_hydrores = np.round(n.storage_units_t.p[n.storage_units[n.storage_units.carrier == 'hydro'].index].sum().sum(),1)
    gen_hydro = gen_hydrores + gen['ror']

    sto_caps = n.storage_units[['p_nom_opt','carrier']].groupby('carrier').sum()
    gen_caps = n.generators[['p_nom_opt','carrier']].groupby('carrier').sum()
    cap_onwind = gen_caps.loc['onwind']
    cap_offwind = gen_caps.loc['offwind-ac'] + gen_caps.loc['offwind-dc']
    cap_solar = gen_caps.loc['solar'] + gen_caps.loc['solar rooftop']
    cap_hydro = gen_caps.loc['ror'] + sto_caps.loc['hydro']
    cap_gas = gen_caps.loc['gas']

    A = n.generators_t.p.sum().sum()
    B = n.storage_units_t.p[n.storage_units.carrier[n.storage_units.carrier == 'hydro'].index ].sum().sum()
    C = (-n.links_t.p1[n.links.carrier[n.links.carrier == 'OCGT'].index]).sum().sum()

    elec_gen_tot = A + B + C

    listt = [energy_storage_cost,
             co2_cap*1e2,
             charge_efficiency,
             discharge_efficiency,
             tau,
             charge_cost,
             discharge_cost,
             trans,
             nnodes,
             tres,
             co2_price,
             system_cost,
             load_shifting*1e2,
             battery_load_shifting*1e2,
             discharge_capacity*1e-3,
             energy_capacity*1e-3,
             charge_capacity*1e-3,
             avail_onwind,
             avail_offwind,
             avail_solar,
             used_onwind,
             used_offwind,
             used_solar,
             gen_gas,
             gen['onwind'],
             gen['offwind'],
             gen['solar'],
             gen_hydro,
             elec_gen_tot,
             cap_onwind.item()/1e3,
             cap_offwind.item()/1e3,
             cap_solar.item()/1e3,
             cap_hydro.item()/1e3,
             cap_gas.item()/1e3]

    listt = pd.DataFrame(listt).round(decimals=3)
    listt = listt[::].append(['-'.join(sector_i)],ignore_index=True)

    if j == 0:
        df = pd.DataFrame(index=np.arange(len(listt)))

    df[j] = listt

    stores = n.stores[n.stores.carrier == 'H2']

    e_cap = stores.e_nom_opt 
    g_cap = n.links[n.links.carrier == 'H2 Fuel Cell'].p_nom_opt
    
    dc_links = n.links.loc[n.links.carrier[n.links.carrier == 'DC'].index]
    kappa_links = dc_links.p_nom_opt
    dc_links_ur = abs(n.links_t.p1[dc_links.index]/kappa_links.sum()*100)

    dc_links_ur_mean_df = pd.DataFrame()
    dc_links_ur_mean_df['ur'] = dc_links_ur.mean()
    dc_links_ur_mean_df['links'] = dc_links[['bus0','bus1']].agg('-DC-'.join, axis=1)
    dc_links_ur_mean_df.set_index('links',drop=True,inplace=True)

    dc_links_ur_var_df = pd.DataFrame()
    dc_links_ur_var_df['ur'] = dc_links_ur.var()
    dc_links_ur_var_df['links'] = dc_links[['bus0','bus1']].agg('-DC-'.join, axis=1)
    dc_links_ur_var_df.set_index('links',drop=True,inplace=True)

    ac_lines = n.lines
    kappa_lines = n.lines.s_nom_opt*(1/np.sqrt(2))
    ac_lines_ur = abs(n.lines_t.p1/(kappa_lines.sum())*100)
    ac_lines_ur_mean_df = pd.DataFrame()
    ac_lines_ur_mean_df['ur'] = ac_lines_ur.mean()
    ac_lines_ur_mean_df['lines'] = ac_lines[['bus0','bus1']].agg('-AC-'.join, axis=1)
    ac_lines_ur_mean_df.set_index('lines',drop=True,inplace=True)

    ac_lines_ur_var_df = pd.DataFrame()
    ac_lines_ur_var_df['ur'] = ac_lines_ur.var()
    ac_lines_ur_var_df['lines'] = ac_lines[['bus0','bus1']].agg('-AC-'.join, axis=1)
    ac_lines_ur_var_df.set_index('lines',drop=True,inplace=True)
    
    df_link[j] = g_cap/1e3 # Discharge power capacity in GW
    df_store[j] = e_cap/1e3 # H2 energy capacity in GWh 
    df_AClines_ur_mean[j] = ac_lines_ur_mean_df['ur']
    df_DClinks_ur_mean[j] = dc_links_ur_mean_df['ur']
    df_AClines_ur_var[j] = ac_lines_ur_var_df['ur']
    df_DClinks_ur_var[j] = dc_links_ur_var_df['ur']
    
    gen1 = {}
    test1 = pd.DataFrame()
    test1[0] = n.generators.carrier
    techs1 = ['onwind','offwind','solar','ror']
    for tech1 in techs1:
        test1[1] = [tech1 in n.generators.carrier[s] for s in range(len(n.generators))]
        gen1[tech1] = n.generators_t.p[n.generators[test1[1]].index].sum()
    gen_gas = -n.links_t.p1[n.links[n.links.carrier == 'OCGT'].index].sum()
    gen_hydrores = n.storage_units_t.p[n.storage_units[n.storage_units.carrier == 'hydro'].index].sum()
    gen_hydro = gen_hydrores + gen1['ror']
    
    storage_discharge_t = -n.links_t.p1[n.links[n.links.carrier == 'H2 Fuel Cell'].index].sum() # MWe (after conversion to electricity)
    load_shifting_g = storage_discharge_t/n.loads_t.p.sum().sum()*1e2
    #load_shifting_g.reindex(load_shifting_g.index + ' loadshift g')
    load_shifting_l = pd.Series(storage_discharge_t.values/n.loads_t.p.sum().values,index=n.loads_t.p.sum().index + ' loadshift l')*1e2
    
    battery_storage_discharge_t = -n.links_t.p1[n.links[n.links.carrier == 'battery discharger'].index].sum()
    battery_load_shifting_g = battery_storage_discharge_t/n.loads_t.p.sum().sum()*1e2
    #battery_load_shifting_g.reindex(battery_load_shifting_g.index + ' battery loadshift g')
    battery_load_shifting_l =  pd.Series(battery_storage_discharge_t.values/n.loads_t.p.sum().values,index=n.loads_t.p.sum().index + ' battery loadshift l')*1e2

    df_onwind[j] = gen1['onwind']
    df_offwind[j] = gen1['offwind']
    df_solar[j] = gen1['solar']
    df_ror[j] = gen1['ror']
    df_hydro[j] = gen_hydrores # gen1['hydro']
    df_gas[j] = gen_gas #1['gas']
    df_battery_loadshift_g[j] = battery_load_shifting_g
    df_battery_loadshift_l[j] = battery_load_shifting_l
    df_storage_loadshift_g[j] = load_shifting_g
    df_storage_loadshift_l[j] = load_shifting_l
    
    j += 1
```

```python
df.index = ['c_hat [EUR/kWh]','co2_cap [%]',
            'eta1 [-]', 'eta2 [-]', 'tau [n_days]',
            'c1','c2','transmission','nnodes','tres',
            'CO2 price [EUR]','c_sys [bEUR]',
            'load_shift [%]','battery_load_shift [%]',
            'G_discharge [GW]','E [GWh]','G_charge [GW]',
            'avail_onwind [MWh]','avail_offwind [MWh]','avail_solar [MWh]',
            'used_onwind [MWh]','used_offwind [MWh]','used_solar [MWh]',
            'gas_gen [MWh]', 'onwind_gen [MWh]', 
            'offwind_gen [MWh]', 'solar_gen [MWh]','hydro_gen [MWh]','tot_gen [MWh]',
            'cap_onwind [GW]','cap_offwind [GW]','cap_solar [GW]',
            'cap_hydro [GW]','cap_gas [GW]',
            'sector']
```

```python
df_storage_loadshift_g.set_index(df_storage_loadshift_g.index + ' loadshift g',inplace=True)
df_battery_loadshift_g.set_index(df_battery_loadshift_g.index + ' loadshift g',inplace=True)
```

```python
df = pd.concat([df.round(decimals=3),
               df_store.round(decimals=3),
               df_link.round(decimals=3),
               df_AClines_ur_mean.round(decimals=3),
               df_DClinks_ur_mean.round(decimals=3),
               df_AClines_ur_var.round(decimals=3),
               df_DClinks_ur_var.round(decimals=3),
               df_onwind.round(decimals=3),
               df_offwind.round(decimals=3),
               df_solar.round(decimals=3),
               df_ror.round(decimals=3),
               df_hydro.round(decimals=3),
               df_gas.round(decimals=3),
               df_battery_loadshift_g.round(decimals=3),
               df_battery_loadshift_l.round(decimals=3),
               df_storage_loadshift_g.round(decimals=3),
               df_storage_loadshift_l.round(decimals=3)])
```
